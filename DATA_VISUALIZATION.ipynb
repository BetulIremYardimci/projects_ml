```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(34)
x = np.arange(20)
y = np.random.normal(10, 1, 20)
z = np.random.normal(10, 2, 20)
```

```python
plt.plot(x, y, z)
```

```python
plt.plot(x,y, color="r")
plt.plot(x,z, color="g");
```

```python
fig = plt.figure(figsize = (10, 6))
ax = fig.add_subplot(131)
ax.set_title("change respective to x")
ax.set_xlabel("values of x")
ax.set_ylabel("values of y")
ax.plot(x,y)

bx = fig.add_subplot(132)
bx.set_title("change respective to y")
bx.set_xlabel("values of x")
bx.set_ylabel("values of y")
bx.plot(x,z, color="y");
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(311)
ax.set_title("change respective to x")
ax.plot(x,y)

bx = fig.add_subplot(312)
bx.set_title("change respective to y")
bx.plot(x,z, color="y");
```

```python
#Feature Engineering
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df['Date']
```

```python
df["day"] = (df['Date']).dt.day
df["month"] = (df['Date']).dt.month
df["year"] = (df['Date']).dt.year
```

```python
df["month_name"] = df['Date'].dt.month_name()
df["weekday"] = df['Date'].dt.day_name()
```

```python
df[["day", "year","month", "month_name", "weekday"]]
```

```python
df["Time"] = pd.to_datetime(df['Time'])
```

```python
df["Hour"] = df["Time"].dt.hour
```

```python
df[["day", "year","month", "month_name", "weekday", "Hour"]]
```
