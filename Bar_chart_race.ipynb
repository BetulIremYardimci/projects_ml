```python
import pandas as pd
```

```python
df = pd.read_csv("/corona_dat.csv")
```

```python
df2 = df[["date", "China", "Italy", "Brazil", "Spain", "US", "United Kingdom", "France", "Turkey"]]
```

```python
df.info()
```

```python
#index yerine değer atama işlemi:
df2.set_index("date", inplace=True)
```

```python
df2_total = df2.cumsum(axis=0)
```

```python
df2_total.head()
```

```python
import bar_chart_race as bcr
```

```python
bcr.bar_chart_race(df2_total, filename="covid19.mp4", figsize=(8, 5), title="Betül - COVID19 VERILERİ")
```
